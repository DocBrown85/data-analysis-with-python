## Setup

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df.head()
```

## Try total_rooms with total_bedrooms

### Simple visualization

```python
sns.relplot(data = df, x = 'total_rooms', y = 'total_bedrooms')
```

### Prepare independent variable

```python
x = np.array(df['total_rooms']).reshape(-1,1)
```

### Prepare dependent variable

```python
y = np.array(df['total_bedrooms']).reshape(-1, 1)
```

### Split train and test data

```python
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
```

### Train the model

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
model = reg.fit(x_train, y_train)
```

```python
model.coef_, model.intercept_
```

### Evaluate the model

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score

y_pred = model.predict(x_test)
mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
mse = mean_squared_error(y_true=y_test,y_pred=y_pred)
rmse = mean_squared_error(y_true=y_test,y_pred=y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('MAE: {} MSE: {} RMSE: {} R^2 {}'.format(mae, mse, rmse, r2))
```

### Visualiza your result

```python
plt.scatter(x_test, y_test)
plt.scatter(x_test, y_pred)
```
